```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
df = {"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(data=dataset,x="x1",y="x2",hue="y")
plt.show()
```

```python
x_a = dataset.iloc[:,:-1]
y_a = dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_a,y_a,test_size=0.2)
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg = BaggingClassifier(estimator=SVC(),n_estimators=30)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train)
```

```python
bg.score(x_test,y_test)
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_test,y_test),sv.score(x_train,y_train)
```

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train,y_train)
rf.score(x_test,y_test),rf.score(x_train,y_train)
```

```python
#Regression
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import LinearRegression
```

```python
bg = BaggingRegressor(estimator =LinearRegression(),n_estimators=100)
bg.fit(x_train,y_train)
bg.score(x_test,y_test)*100,bg.score(x_train,y_train)*100
```

```python
rf = RandomForestRegressor(n_estimators=55)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)*100,rf.score(x_train,y_train)*100
```

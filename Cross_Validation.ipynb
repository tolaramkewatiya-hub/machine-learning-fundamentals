```python
import pandas as pd
from sklearn import datasets,linear_model
```

```python
wine = datasets.load_wine()
print(wine.keys())
```

```python
x = wine.data
y = wine.target
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
p=cross_val_score(LinearRegression(),x,y,cv=5)
p.sort()
p
```

```python
new_dat = pd.DataFrame(wine.data).head(10)
new_dat
```

```python
x_new = new_dat
y_new = wine.target[:10]
```

```python
from sklearn.model_selection import LeaveOneOut,LeavePOut,KFold,StratifiedKFold
lo = LeaveOneOut()
for train,test in lo.split(x_new,y_new):
 print(train,test)
```

```python
lo = LeavePOut(p=2)
for train,test in lo.split(x_new,y_new):
 print(train,test)
```

```python
lo = KFold(n_splits=5)
for train,test in lo.split(x_new,y_new):
 print(train,test)
```

```python
lo = StratifiedKFold(n_splits=5)
for train,test in lo.split(x_new,y_new):
 print(train,test)
```
